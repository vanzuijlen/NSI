### Le problème :écrire une fonction récursive qui calcule la somme des chiffres d'un entier strictement positif

ce qu'il faut faire : somme_des_chiffres(12345) renvoie $1+2+3+4+5=15$

l'idée :

somme_des_chiffres(12345)=somme_des_chiffres(1234) + 5

de plus on sait que :

```python
print(12345//10) # renvoie 1234
print(12345%10)  # renvoie 5
```

```python
def somme_des_chiffres(n):
    assert(type(n)==int),"n doit être un entier"
    if n<10:
        return n
    else:
        return n%10 + somme_des_chiffres(n//10)
somme_des_chiffres(12345)
```

## Les piles

### représenter une pile avec une classe

il faut les méthodes:

* empiler
* depiler
* est_vide
* création d'une méthode qui donne la taille de la liste

```python
# création de la classe pile
class Pile:
    def __init__(self):
        self.pile=[]
    def est_vide(self):
        return self.pile==[]
    def empiler(self,x):
        self.pile.append(x)
    def depiler(self):
        assert not self.est_vide(),"la pile est vide"
        self.pile.pop()
    def taille(self):
        return len(self.pile)
p=Pile()
p.empiler(2)
p.empiler(3)
print(p.taille())
p.depiler()
p.depiler()
#p.depiler()
```

$ ((()())$ est mal parenthesée

```python
def parenthesage(expression):
    p=Pile()
    for car in expression:
        if car=="(":
            p.empiler(car)
        if car==")":
            if p.est_vide()==True:
                return False
            else:
                p.depiler()
    if p.est_vide()==True:
        return True
    else:
        return False
expression1="(2x(3y-1)-3(x+3)"
expression2="((())"

print(parenthesage(expression1))
print(parenthesage(expression2))
```

# Implémenter une classe pile à partir d'une liste chaînée ( cellule )

La classe cellule crée un objet Cellule qui contient -->  ( tête, queue )  ==  ( contenu, suivante )

La classe pile, permet d'empiler une cellule (x, suivante ==> la cellule active devient la suivante et x le contenu)

Lorsque l'on dépile, on récupère le contenu, et le contenu prend la valeur de la suivante

Un plus pour l'affichage sous forme de pile

```python
class Cellule :
    
    def __init__(self, contenu, suivante):
        self.contenu = contenu
        self.suivante = suivante
        


class Pile:
    def __init__(self):
        self.data = None
    
    def est_vide(self):
        return self.data == None
    
    def empile(self, x):
        self.data = Cellule(x,self.data)
    
    def depile(self):
        v = self.data.contenu #on récupère la valeur à renvoyer
        self.data = self.data.suivante  # on supprime la 1ère cellule  
        return v
    
    def __str__(self):
        s = ""
        c = self.data
        if c==None:
            return "| |"
        while c != None :
            s += "|"+str(c.contenu)+"|"+"\n"
            c = c.suivante
        return s


p = Pile()
p.empile(5)
p.empile(3)
p.empile(7)
p.empile(8)
print(p)
q=Pile()
print(q)
```
