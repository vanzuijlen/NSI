# Les files

```
def file():
    #retourne une file vide
    return []

def vide(f):
    ''' renvoie True si la file est vide
     False sinon'''
    return f==[]

def enfiler(f,x):
    #ajoute x à la file f"
    return f.append(x)


def defiler(f):
    '''
    enlève et renvoie le premier élément de la file
    '''
    assert not vide(f), "file vide"
    return f.pop(0)

def taille(f):
    return len(f)

def sommet(f):
    return f[0]
```

## Avec une classe

```
class File:
    ''' classe File
    création d'une instance File avec une liste
    '''
    def __init__(self):
        "Initialisation d'une file vide"
        self.L=[]

    def vide(self):
        "teste si la file est vide"
        return self.L==[]

    def defiler(self):
        "défile"
        assert not self.vide(), "file vide"
        return self.L.pop(0)

    def enfiler(self,x):
        self.L.append(x)

    def taille(self):
        return len(self.L)

    def sommet(self):
        return self.L[0]

    def suppr_dernier(self):
        self.L.pop(-1)
        
    # un plus pour un affichage de la file avec print
    def __str__(self):       
        s = "|"            
        for k in self.L :
            s ="|"+ str(k)+s 
        return s
f=File()
for i in range(5):
    f.enfiler(i)
print(f)
```

# Exercices

```
class Pile:
    ''' classe Pile '''
    def __init__(self):
        "Initialisation d'une pile vide"
        self.L=[]

    def vide(self):
        "teste si la pile est vide"
        return self.L==[]

    def depiler(self):
        "dépile"
        assert not self.vide(),"Pile vide"
        return self.L.pop()

    def empiler(self,x):
        "empile"
        self.L.append(x)

    def taille(self):
        return(len(self.L))

    def sommet(self):
        assert not self.vide(),"Pile vide"
        return self.L[-1]

    def suppr_premier(self):
        self.L.pop(0)

    def __str__(self):       
        s = ""            
        for k in range(self.taille()-1,-1,-1) :
            s =s + "|"+ str(self.L[k])+"|"+"\n"
        return s
p=Pile()
for i in range(5):
    p.empiler(i)
print(p)
```

### Ex4

On dispose d’une file, écrire une fonction qui renvoie la file inversée (l’élément de la tête sera situé à la queue et ainsi de suite ) On utilisera une pile et seulement les méthodes associées aux piles et aux files

```
def renverse_file(f):
    pile=Pile()
    while f.vide()==False:
        pile.empiler(f.defiler())
    while pile.vide()==False:
        f.enfiler(pile.depiler())
    return f
f=File()
for i in range(20):
    f.enfiler(i)
print(f)
print(renverse_file(f))
```

### Ex 6

On dispose d’une file contenant des entiers, écrire une fonction qui renvoie une file où on aura séparé les nombres pairs des impairs

```
def pair_impair_file(f):
    f1=File()
    f2=File()
    while f.vide()==False:
        el=f.defiler()
        if el%2==0:
            f1.enfiler(el)
        else:
            f2.enfiler(el)
    while f1.vide()==False:
        f.enfiler(f1.defiler())
    while f2.vide()==False:
        f.enfiler(f2.defiler())
    return f
f=File()
for i in range(20):
    f.enfiler(i)
print(f)
print(pair_impair_file(f))
```

### Ex 7

même chose avec des piles

```
def pair_impair_pile(p):
    p1=Pile()
    p2=Pile()
    while p.vide()==False:
        el=p.depiler()
        if el%2==0:
            p1.empiler(el)
        else:
            p2.empiler(el)
    while p1.vide()==False:
        p.empiler(p1.depiler())
    while p2.vide()==False:
        p.empiler(p2.depiler())
    return p
p=Pile()
for i in range(10):
    p.empiler(i)
print(p)
print(pair_impair_pile(p))
```

### Ex 8

On dispose d’une pile contenant des entiers, écrire une fonction ou une méthode qui renvoie une pile où l’on aura supprimé le premier élément entré

```
def enleve_premier(p):
    f=File()
    p1=Pile()
    while p.vide()==False:
        f.enfiler(p.depiler())
    f.defiler()
    while f.vide()==False:
        p1.empiler(f.defiler())
    while p1.vide()==False:
        p.empiler(p1.depiler())
    return p
p=Pile()
for i in range(10):
    p.empiler(i)
print(p)
print(enleve_premier(p))
```
