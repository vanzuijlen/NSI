```python
def creation_arbre(r,profondeur):
    ''' r : la racine (str ou int). la profondeur de l'arbre (int)'''
    Arbre = [r]+[None for i in range(2**(profondeur+1)-2)]
    return Arbre
    
def insertion_noeud(arbre,n,fg,fd):
    '''Insére les noeuds et leurs enfants dans l'arbre'''
    indice = arbre.index(n)
    arbre[2*indice+1] = fg
    arbre[2*indice+2] = fd
# création de l'arbre    
arbre = creation_arbre("r",5)
# ajout des noeuds par niveau de gauche à droite   
insertion_noeud(arbre,"r","a","b")   
insertion_noeud(arbre,"a","c","d")
insertion_noeud(arbre,"b","e","f")
insertion_noeud(arbre,"c",None,"h")
insertion_noeud(arbre,"d","i","j")
insertion_noeud(arbre,"e","k",None)
insertion_noeud(arbre,"f",None,None)
insertion_noeud(arbre,"h",None,None)
insertion_noeud(arbre,"i",None,None)
insertion_noeud(arbre,"j","m",None)
insertion_noeud(arbre,"k",None,None)
insertion_noeud(arbre,"m",None,None)
#pour vérifier
print(len(arbre))
print(arbre)
```

### La fonction parent

```python
def parent(arbre,p):
    if p in arbre:
        indice = arbre.index(p)
        if indice%2 == 0:
            return arbre[(indice-2)//2]
        else:
            return arbre[(indice-1)//2]
        
print(parent(arbre,'r'))
print(parent(arbre,'c'))
```

```python
# les fonctions qui renvoie les fils gauche et droit s'ils existent

def fils_g(noeud):
    n=arbre.index(noeud)
    fg=arbre[2*n+1]
    if fg!=None:
        return fg

print(fils_g('j'))

def fils_d(noeud):
    n=arbre.index(noeud)
    fd=arbre[2*n+2]
    if fd!=None:
        return fd

print(fils_d('j'))
```

```python
# Une fonction qui retourne vrai si le noeud est la racine de l’arbre.

def est_racine(noeud):
    if parent(arbre,noeud)==None:
        return True
    else:
        return False
print(est_racine('r'))
print(est_racine('j'))
```

```python
# Une fonction qui retourne vrai si le noeud est une feuille

def est_feuille(noeud):
    if fils_g(noeud)==None and fils_d(noeud)==None :
        return True
    else:
        return False
    
print(est_feuille('j'))
print(est_feuille('k'))
        
```

```python
#  Une fonction qui retourne vrai si le noeud a un frère gauche ou droit

def frere(noeud):
    if noeud in arbre:
        n=arbre.index(noeud)
        if n%2==0: # c'est un fils droit
            k=(n-2)//2 # indice du parent
            if arbre[2*k+1]!=None:
                return True
            else:
                return False
        else: # c'est un fils gauche
            k=(n-1)//2
            if arbre[2*k+2]!=None:
                return True
            else:
                return False
print(frere('i'))
print(frere('m'))
```

## Représentation de l'arbre avec une liste de liste

**La structure de arbre1 est une liste de liste construite à partir des noeuds qui sont des dictionnaires**

On peut très bien se contenter de 'racine' pour représenter l'arbre

```python
def noeud(nom, fg = None, fd = None) :
    return {'racine': nom, 'fg' : fg, 'fd': fd}
# création des noeuds
k = noeud('k')
f = noeud('f')
e = noeud('e', k, None)
b = noeud('b', e, f)
m = noeud('m')
j = noeud('j', m, None)
i = noeud('i')
d = noeud('d', i, j)
h = noeud('h')
c = noeud('c', None, h)
a = noeud('a', c, d)
racine = noeud('r', a, b)
# création de l'arbre
def construit(arbre) :
    if arbre == None :
        return []
    else:
        return [arbre['racine'],construit(arbre['fg']),construit(arbre['fd'])]
arbre1=construit(racine)
print(arbre1)
```

## Le parcours en largeur :

```python
# la classe File
class File:
    ''' classe File
    création d'une instance File avec une liste
    '''
    def __init__(self):
        "Initialisation d'une file vide"
        self.L=[]

    def vide(self):
        "teste si la file est vide"
        return self.L==[]

    def defiler(self):
        "défile"
        assert not self.vide(), "file vide"
        return self.L.pop(0)

    def enfiler(self,x):
        self.L.append(x)
```

```python
print(arbre1[0])
print(arbre1[1])
print(arbre1[2])
```

```python
def parcours_largeur(arbre):
    f=File()
    liste=[]
    f.enfiler(arbre)
    while not f.vide():
        tmp=f.defiler()
        liste.append(tmp[0])
        #print(liste)
        if tmp[1]!=[]:
            f.enfiler(tmp[1])
        if tmp[2]!=[]:
            f.enfiler(tmp[2])
    return liste

print(parcours_largeur(arbre1))
```

## Les parcours en profondeurs

```python
# parcours prefixe

def parcoursPrefixe(arbre):
    if arbre!=[]:
        print(arbre[0],end=" , " )
        parcours_prefixe(arbre[1])
        parcours_prefixe(arbre[2])
parcoursPrefixe(arbre1)
```

```python
# parcours infixe

def parcoursInfixe(arbre) :
    if(arbre != []) :
        parcoursInfixe(arbre[1])
        print(arbre[0], end = ', ')
        parcoursInfixe(arbre[2])
parcoursInfixe(arbre1)
```

```python
# parcours suffixe ou Postfixe

def parcoursPostfixe(arbre) :
    if(arbre != []) :
        parcoursPostfixe(arbre[1])
        parcoursPostfixe(arbre[2])
        print(arbre[0], end = ', ')
parcoursPostfixe(arbre1)
```

# Représentation d'un arbre avec une classe

```python
class Noeud:

    # Le constructeur
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
    # Méthode qui permet d'afficher la valeur
    # de la racine avec la fonction print
    def __str__(self):
        return str(self.value)
    
    def est_feuille(self):
        return self.left==None and self.right==None
    
    def hauteur(self):
        h1=0
        h2=0
        if not self:
            return -1
        else:
            if self.left:
                h1 = 1 + self.left.hauteur()
            if self.right:
                h2 = 1 + self.right.hauteur()
            return max(h1, h2)
        
    def parcours_infixe(self):
        if self:
            if self.left:
                self.left.parcours_infixe()
            print(self.value,end=' ')
            if self.right:
                self.right.parcours_infixe()

    def parcours_prefixe(self):
        if self:
            print(self.value,end=' ')
            if self.left:
                self.left.parcours_prefixe()
            if self.right:
                self.right.parcours_prefixe()

    def parcours_suffixe(self):
        if self:
            if self.left:
                self.left.parcours_suffixe()
            if self.right:
                self.right.parcours_suffixe()
            print(self.value,end=' ')

             

k = Noeud('k')
f = Noeud('f')
e = Noeud('e', k, None)
b = Noeud('b', e, f)
m = Noeud('m')
j = Noeud('j', m, None)
i = Noeud('i')
d = Noeud('d', i, j)
h = Noeud('h')
c = Noeud('c', None, h)
a = Noeud('a', c, d)
arbre3 = Noeud('r', a, b)

print(arbre3) # affiche r 
print(arbre3.left,arbre3.right)
print(a.left)
print(arbre3.left.left)
print(j.est_feuille())
print(h.est_feuille())
print(arbre3.hauteur())
arbre3.parcours_prefixe()
print("")
arbre3.parcours_infixe()
print("")
arbre3.parcours_suffixe()
```

# Le TP : évaluer une expression arithmétique

```python
####################################################################
#                 evaluation d'expression
####################################################################

def evalue(arbre):
    if arbre.est_feuille():
        res=arbre.value
    elif arbre.value=='+':
        res= evalue(arbre.left)+evalue(arbre.right)
    elif arbre.value=='-':
        res= evalue(arbre.left)-evalue(arbre.right)
    elif arbre.value=='*':
        res= evalue(arbre.left)*evalue(arbre.right)
    elif arbre.value=='/':
        res= evalue(arbre.left)/evalue(arbre.right)
    else:
        res=None
    return res
# affichage parenthésée
def printexp(tree):
    sVal = ""
    if tree:
        sVal = '(' + printexp(tree.left)
        sVal = sVal + str(tree.value)
        sVal = sVal + printexp(tree.right)+')'
    return sVal






####################################################################
#                       Création de l'arbre                        #
####################################################################
f=Noeud(7)
e=Noeud(4)
d=Noeud('-',e,f)
c=Noeud(3)
b=Noeud('/',c,d)
a=Noeud(2)
arbre = Noeud('+', a, b)
#graphicarbre(arbre)
arbre.parcours_infixe()
print('\n')
arbre.parcours_prefixe()
print('\n')
arbre.parcours_suffixe()
print('\n')
#res=0
print(evalue(arbre))
print(f.est_feuille())
print(printexp(arbre))
#graphicarbre(arbre)
```
